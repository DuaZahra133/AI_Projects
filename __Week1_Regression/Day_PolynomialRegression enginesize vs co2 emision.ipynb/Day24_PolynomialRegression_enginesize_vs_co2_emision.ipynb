# Day 24: Polynomial Regression
** Project:** Engine Size vs CO2 Emission

**Objective:** Fit a polynomial regression model to predict CO2 emissions based on engine size. Compare linear and polynomial regression.


```python
# ============================
# 1️⃣ Import Libraries
# ============================
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
```

## 2️⃣ Dataset
I am using using a small dataset showing Engine Size (in liters) vs CO2 Emission (g/km).  
This dataset demonstrates a curved relationship suitable for polynomial regression.


```python
# ============================
# 2️⃣ Dataset Creation
# ============================

# Engine Size (L) vs CO2 Emission (g/km)
X = np.array([1.0, 1.2, 1.4, 1.6, 1.8, 2.0, 2.2, 2.4]).reshape(-1,1)
y = np.array([120, 130, 145, 160, 180, 200, 230, 260])

# Scatter plot of the dataset
plt.scatter(X, y, color='blue')
plt.title("Engine Size vs CO2 Emission")
plt.xlabel("Engine Size (L)")
plt.ylabel("CO2 Emission (g/km)")
plt.show()
```

## 3️⃣ Linear Regression for Comparison
I will first fit a linear regression model to see how it performs on this curved data.


```python
# ============================
# Linear Regression
# ============================
lin_reg = LinearRegression()
lin_reg.fit(X, y)
y_lin_pred = lin_reg.predict(X)

plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_lin_pred, color='red', label='Linear Regression')
plt.title("Linear Regression Fit")
plt.xlabel("Engine Size (L)")
plt.ylabel("CO2 Emission (g/km)")
plt.legend()
plt.show()

# Evaluate Linear Regression
r2_lin = r2_score(y, y_lin_pred)
print(f"R² score for Linear Regression: {r2_lin:.2f}")
```

## 4️⃣ Polynomial Regression (Degree 2)
Now I will fit a polynomial regression model of degree 2 to better capture the curved relationship.


```python
# Polynomial Regression Degree 2
poly2 = PolynomialFeatures(degree=2)
X_poly2 = poly2.fit_transform(X)

poly_reg2 = LinearRegression()
poly_reg2.fit(X_poly2, y)
y_poly2_pred = poly_reg2.predict(X_poly2)

plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_poly2_pred, color='red', label='Polynomial Regression (Degree 2)')
plt.title("Polynomial Regression - Degree 2")
plt.xlabel("Engine Size (L)")
plt.ylabel("CO2 Emission (g/km)")
plt.legend()
plt.show()

r2_poly2 = r2_score(y, y_poly2_pred)
print(f"R² score for Polynomial Regression (Degree 2): {r2_poly2:.2f}")
```

## 5️⃣ Polynomial Regression Experiment
Now we will try higher-degree polynomials (3 & 4) to see how the curve fits the data and check for overfitting.


```python
degrees = [3, 4]
for degree in degrees:
    poly = PolynomialFeatures(degree=degree)
    X_poly = poly.fit_transform(X)
    model = LinearRegression()
    model.fit(X_poly, y)
    y_pred = model.predict(X_poly)

    plt.scatter(X, y, color='blue', label='Actual Data')
    plt.plot(X, y_pred, color='red', label=f'Polynomial Regression (Degree {degree})')
    plt.title(f'Polynomial Regression - Degree {degree}')
    plt.xlabel('Engine Size (L)')
    plt.ylabel('CO2 Emission (g/km)')
    plt.legend()
    plt.show()

    print(f"Degree {degree}: R² = {r2_score(y, y_pred):.2f}")
```

## 6️⃣ Notes & Summary

- Linear Regression fails to capture the curved relationship.
- Polynomial Regression fits curves and improves R² score.
- Degree selection is important: too low = underfitting, too high = overfitting.
- Visualization helps to see the model fit.
- Key Python Functions: `PolynomialFeatures`, `LinearRegression`, `fit_transform`, `predict`.
